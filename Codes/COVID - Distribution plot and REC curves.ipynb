#### REC Curve

```python
df = pd.read_csv('Accuracy_deviation.csv')
df 
```

```python
plt.rcParams['font.family'] = 'sans-serif'
```

```python

plt.plot(df['Deviation'], df['Accuracy_knn'], label='KNN = 0.949')
plt.plot(df['Deviation'], df['Accuracy_rf'], label='RF = 0.949')
plt.plot(df['Deviation'], df['Accuracy_svm'], label='SVM = 0.950')
plt.plot(df['Deviation'], df['Accuracy_ensemble'], label='Ensemble = 0.952')
plt.plot([0, 1], [0, 1], ls='--', color='black')

plt.xlabel('Deviation', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)

plt.legend(title = 'AUC', fontsize = 10)

plt.grid()
plt.savefig(r'F:\Ishani\Modeling\Images\REC_AUC_plot.tiff', dpi=600)
```
